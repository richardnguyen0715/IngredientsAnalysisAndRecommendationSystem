```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
ingredients_df = pd.read_csv("../Assert/ingredients.csv",sep=",")
ingredients_df.head()
```

```python
def dishesRecommenderKNN(ingredients_input: pd.DataFrame):
    # Tách nguyên liệu (input) và tên món ăn (label)
    ingredients = ingredients_df.iloc[:,1:]
    print(ingredients.shape)
    ingredients = ingredients[~(ingredients == 0).all(axis=1)]
    print(ingredients.shape)
    
    dishes = ingredients_df.iloc[:,0]

    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(ingredients)  

    # Khởi tạo và huấn luyện mô hình KNN
    k = 5
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(ingredients, dishes)

    _,indices = knn.kneighbors(ingredients_input)
    print(_,indices)
    predict_dishes = dishes.iloc[indices[0][:k]].values

    return predict_dishes
```

```python
# TEST
input = ['active yeast']
ingredients_list = ingredients_df.columns
all_ingredients_size = ingredients_list.size 

input_list = []
for i in range(1, all_ingredients_size):
    if ingredients_list[i] in input:
        input_list.append(1)
    else:
        input_list.append(0)

input_df = pd.DataFrame([input_list], columns=ingredients_list[1:])
```

```python
predict = dishesRecommenderKNN(input_df)
print(predict)
```
